```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'Iris.csv'
iris_df = pd.read_csv(file_path)
iris_df
```

```python
iris = sns.load_dataset('iris')
```

```python
# Bar Plot
plt.figure(figsize=(10, 6))
iris.groupby('species').mean().plot(kind='bar', ax=plt.gca())
plt.title('Mean Values of Iris Dataset Features by Species')
plt.ylabel('Mean Value')
plt.show()
```
